**LeNet** Structure:
1. Input Layer: Output shape=32x32x1
2. Conv2D[6 filters of size=5x5, stride=1,padding=0] ,Output shape=28x28x6
3. Average Pooling [Stride=2,padding=0],              Output shape=14x14x6
4. Conv2D[16 filters of size=5x5, stride=1,padding=0],Output shape=10x10x16
5. Average Pooling [Stride=2,padding=0],              Output shape=5x5x16
6. Conv2D[120 filters of size=5x5, stride=1,padding=0],Output shape=1x1x120
7. Linear1 Layer                                      ,Output shape=120
8. Linaer2 Layer                                      ,Output shape=84
9. Final Linear Layer                                 ,Output shape=10

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adadelta
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
img_rows=x_train[0].shape[0]  #storing no of rows and columns
img_cols=x_train[1].shape[0]
x_train=x_train.reshape(x_train.shape[0],img_rows,img_cols,1) #adding 4th dimmension
x_test=x_test.reshape(x_test.shape[0],img_rows,img_cols,1)
input_shape=(img_rows,img_cols,1)  #storing shape of an input
x_train=x_train.astype('float32')  #Changing our image type to float
x_test=x_test.astype('float32')
x_train=x_train/255                #Normalizing
x_test=x_test/255
y_train=to_categorical(y_train)    #One hot encoding
y_test=to_categorical(y_test)
num_classes=y_test.shape[1]
num_pixels=x_train.shape[1]*x_train.shape[2]
```

```python
model=Sequential()

model.add(Conv2D(6,(5,5),padding='same',input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))

model.add(Conv2D(16,(5,5),padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))

model.add(Conv2D(120,(5,5),padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))

model.add(Flatten())
model.add(Dense(120))
model.add(Activation('relu'))

model.add(Dense(84))
model.add(Activation('relu'))

model.add(Dense(num_classes))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',optimizer=Adadelta(),metrics=['accuracy'])
model.summary()
```

```python
history=model.fit(x_train,y_train,batch_size=16,epochs=50,validation_data=(x_test,y_test),shuffle=True)
model.save('mnist_LeNet.h5')
scores=model.evaluate(x_test,y_test,verbose=1)
print('Test Loss:', scores[0])
print('Test Accuracy:', scores[1])
```

**ALEXNET**

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adadelta
```

```python
(x_train,y_train),(x_test,y_test)=cifar10.load_data()
print('x_train shape=', x_train.shape)
print(x_train.shape[0],'train examples')
print(x_test.shape[0],'test examples')
num_classes=10
y_train=to_categorical(y_train,num_classes)
y_test=to_categorical(y_test,num_classes)
```

```python
l2_reg=0
model=Sequential()

model.add(Conv2D(96,(11,11),input_shape=x_train.shape[1:],padding='same',kernel_regularizer=l2(l2_reg))) #1st conv
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(256,(5,5),padding='same')) #2nd conv
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(ZeroPadding2D((1,1)))   #3rd conv
model.add(Conv2D(512,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(ZeroPadding2D((1,1)))  #4th conv
model.add(Conv2D(1024,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))

model.add(ZeroPadding2D((1,1)))  #5th conv
model.add(Conv2D(1024,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())         #1st fully connected layer
model.add(Dense(3072))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(4096))       #2nd fully connected layer
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(num_classes))
model.add(BatchNormalization())
model.add(Activation('softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer=Adadelta(),metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,batch_size=128,epochs=50,validation_data=(x_test,y_test),shuffle=True)
model.save('cifar10_alexnet.h5')
scores=model.evaluate(x_test,y_test,verbose=1)
print('Test loss:',scores[0])
print('Test accuracy',scores[1])
```
